```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
%matplotlib inline
```

```python
# Import the dataset from keras
(X_train, y_train),(X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Check shape of the dataset
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# cehck one data matrix dimentions
X_train[0].shape
```

```python
# Print 20 images from the dataset
for i in X_train[:20]:
    plt.matshow(i)
```

```python
# Normalize the dataset
X_train = X_train/255
X_test = X_test/255

# Flatten the dataset in order to compute for model building.
X_train_flatten = X_train.reshape(len(X_train), 28*28)
X_test_flatten = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flatten.shape,y_train.shape
```

```python
# convert the y_train to catagorical in order to run catagorical_crossentropy
y_train = keras.utils.to_categorical(y_train) # this is important
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train_flatten,y_train,epochs=5)
```

```python
y_test = keras.utils.to_categorical(y_test)
```

```python
model.evaluate(X_test_flatten,y_test)
```
